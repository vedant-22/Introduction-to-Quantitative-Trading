```python
import yfinance as yf

def stock(str1):
    inp1 = yf.download(str1 , start="2022-01-01", end="2023-01-01")['Open']
    inp2 = yf.download(str1, start="2022-01-01", end="2023-01-01")['Close']
    inp3 = yf.download(str1, start="2022-01-01", end="2023-01-01")['High']
    inp4 = yf.download(str1, start="2022-01-01", end="2023-01-01")['Low']
    
    oc = inp1/inp2
    ol = inp1/inp4
    hl = inp3/inp4
    hc = inp3/inp2
    
    a=b=c=d=a1=a2=a3=b1=b2=b3=c1=c2=c3=d1=d2=d3=0
    
    if(oc.iloc[0] == 0.35):
        a=a+1

    if(oc.iloc[0] == 0.5):
        b=b+1

    if(oc.iloc[0] == 1):
        c=c+1

    if(oc.iloc[0] == 3.5):
        d=d+1

    if(ol.iloc[0] == 0.35):
        a1=a1+1

    if(ol.iloc[0] == 0.5):
        b1=b1+1

    if(ol.iloc[0] == 1):
        c1=c1+1

    if(ol.iloc[0] == 3.5):
        d1=d1+1 
        
    if(hl.iloc[0] == 0.35):
        a2=a2+1

    if(hl.iloc[0] == 0.5):
        b2=b2+1

    if(hl.iloc[0] == 1):
        c2=c2+1

    if(hl.iloc[0] == 3.5):
        d2=d2+1

    if(hc.iloc[0] == 0.35):
        a3=a3+1

    if(hc.iloc[0] == 0.5):
        b3=b3+1

    if(hc.iloc[0] == 1):
        c3=c3+1

    if(hc.iloc[0] == 3.5):
        d3=d3+1  
        
        
    print("The no. of times Open/Close ratio becomes equal to 0.35, 0.5, 1, 3.5 respetively is:", a, b, c, d)
    print("The no. of times Open/Low ratio becomes equal to 0.35, 0.5, 1, 3.5 respetively is:", a1, b1, c1, d1)
    print("The no. of times High/Low ratio becomes equal to 0.35, 0.5, 1, 3.5 respetively is:", a2, b2, c2, d2)
    print("The no. of times High/Close ratio becomes equal to 0.35, 0.5, 1, 3.5 respetively is:", a3, b3, c3, d3)
    

stock('AAPL')
stock('AMZN')
stock('GOOGL')
stock('MSFT')
stock('TSLA')

        
        
        
    
    
    
    


```

```python
import matplotlib.pyplot as plt
%matplotlib inline

tickers_list = ['AAPL', 'AMZN']

data = yf.download(tickers_list, start="2022-01-01", end="2023-01-01")['Close']
# Plot all the close prices
plt.figure(figsize=(10,6))
data.plot()

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Closing Price vs Year for AAPL and AMZN", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Closing Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
